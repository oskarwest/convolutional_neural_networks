```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img=io.imread('/kaggle/input/cnn-data-sources/yo.png')
img_gray=color.rgb2gray(img)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray)
```

```python
kernel=np.array([[1,1,1],
                [0,0,0],
                [-1,-1,-1]]
               )
```

```python
img_new=nd.convolve(img_gray,kernel)
```

```python
fig, ax=plt.subplots(1,2, figsize=(15,10))
ax[0].imshow(img_gray, cmap=plt.cm.gray)
ax[0].set_title('original')
ax[0].axis('off')

ax[1].imshow(img_new, cmap=plt.cm.gray)
ax[1].set_title('convolution')
ax[1].axis('off')
```
