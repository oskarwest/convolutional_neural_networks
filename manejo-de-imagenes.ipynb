```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im = io.imread('/kaggle/input/cnn-data-sources/Mara.jpeg')
im.shape
```

```python
plt.imshow(im)
```

```python
r = im[:,:,0]
g = im[:,:,1]
b = im[:,:,2]
```

```python
plt.imshow(r.T, cmap='gray')
```

```python
r.shape
aux_dim=np.zeros([800,600])
```

```python
im_neg_pos=255-im
im32=(im//32)*32
im128=(im//128)*128
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(1,4)
ax1.imshow(im)
ax1.axis('off')
ax1.set_title('original')
ax2.imshow(im_neg_pos)
ax2.axis('off')
ax2.set_title('positivo a negativo')
ax3.imshow(im32)
ax3.axis('off')
ax3.set_title('baja resolucion')
ax4.imshow(im128)
ax4.axis('off')
ax4.set_title('baja resolucion')

plt.tight_layout()
plt.show()
```
